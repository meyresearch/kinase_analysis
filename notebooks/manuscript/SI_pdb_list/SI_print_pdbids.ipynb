```python
from pathlib import Path
from natsort import natsorted
```

```python
abl_dir = Path('/arc/human_abl')
abl_crystals = list(abl_dir.glob('*.pdb'))
pdb_ids = [p.stem.split('_')[3] for p in abl_crystals]
string = ", ".join(natsorted(set(pdb_ids))) 
string
```
